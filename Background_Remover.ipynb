```python
from PIL import Image
import os, os.path

#Path from the Images we want to convert
path = "./"

#Path where the Images without Background will be stored
destinationPath = "./NoBackground/"

#Usefull prints to know if you are in the correct path, should show your Images 
#names and the number of total elements in the Folder
print("Elements",os.listdir(path))
print("Total Elements",len(os.listdir(path)))

valid_images = [".jpg",".gif",".png",".tga"]

#Loop that gets all the elements in de origin path one by one
for item in os.listdir(path):
    #The extension will be checked, if not in "valid_images" this element will be ignored
    ext = os.path.splitext(item)[1]
    if ext.lower() in valid_images:
        #If its an Image its Background will be removed
        img = Image.open(item)
        img = img.convert("RGBA")
        datas = img.getdata()
        
        newData = []
        for item in datas:
            if item[0] == 255 and item[1] == 255 and item[2] == 255:
                newData.append((255, 255, 255, 0))
            else:
                if item[0] > 150:
                    newData.append((0, 0, 0, 255))
                else:
                    newData.append(item)
                    print(item)
        
        #Number of elements in Destination folder, used to calculate the name for the new stored image
        print("Lenght Destination folder ", len(os.listdir(destinationPath)))
        
        #Creating name
        ext=str(len(os.listdir(destinationPath)))+".png"
        
        #Image saved in the destination path with the name and extension previously created
        img.save(os.path.join(destinationPath, ext))
    

```
